```python
import re

file_name = "Inputs\\input_05.txt"

updates = []

rules = []

with open(file_name, 'r') as file:

    lines = file.readlines()

for line in lines:

    if '|' in line:

        cleaned_line = re.sub(r'\s', '', line)
        rules.append(cleaned_line)

    else:

        if len(line) > 1:

            cleaned_line = re.sub(r'\s', '', line)
            updates.append(cleaned_line)

rules
```

```python
#rules
rule_list = []

for rule in rules:

    rule_list.append(rule.split('|'))

rule_list
```

```python
violations = []

for update in updates:

    update = update.split(',')

    for rule in rule_list:

        x = rule[0]

        y = rule[1]

        if x in update and y in update:

            if update.index(x) > update.index(y):

                violations.append(update)

unique_violations = []

for violation in violations:

    if violation not in unique_violations:

        unique_violations.append(violation)


```

```python
correct_updates = []

for update in updates:

    update = update.split(',')

    if update not in unique_violations:

        correct_updates.append(update)

odd_updates_mid_values = []

even_updates_mid_values = []

for update in correct_updates:

    if len(update) % 2 != 0:

        mid_value = update[len(update) // 2]

        odd_updates_mid_values.append(mid_value)

    if len(update) % 2 == 0:

        middle_left = update[len(update) // 2 - 1]
        middle_right = update[len(update) // 2]
        average_middle = (middle_left + middle_right) / 2

        even_updates_mid_values.append(average_middle)

list_ = [int(x) for x in odd_updates_mid_values]
sum(list_)
```

```python
import re

import numpy as np

file_name = "Inputs\\input_06.txt"

with open(file_name, 'r') as file:

    lines = file.readlines()

matrix_list = []

for line in lines:

    line = re.sub(r'\s', '', line)

    points = []

    for point in line:

        points.append(point)

    matrix_list.append(points)

matrix = np.array(matrix_list)

matrix

# Export to a CSV file
np.savetxt("output.csv", matrix, delimiter=",", fmt='%s', header="Col1,Col2,Col3", comments='')

print("Array exported to 'output.csv'.")
```

```python
# location of '^'

for i,row in enumerate(matrix):

    guard = '^'

    if guard in row:

        start_y = i

        column = row.tolist().index(guard)

        start_x = column

        print(matrix[start_y,start_x])

        print(start_y, start_x)

```

```python
def step_function(y, x, direction):

    step = []

    if direction == 'up':

        next_direction = 'right'

        while matrix[y,x] != '#':

            step.append([y,x])

            y = y - 1
        
        res_dict = {
            'y': y+1,
            'x': x,
            'len_steps': len(step),
            'nxt_dir': next_direction,
            'steps': step
        }

    if direction == 'right':

        next_direction = 'down'

        while matrix[y,x] != '#':

            step.append([y,x])

            x = x + 1
        
        res_dict = {
            'y': y,
            'x': x-1,
            'len_steps': len(step),
            'nxt_dir': next_direction,
            'steps': step
        }

    if direction == 'down':

        next_direction = 'left'

        while matrix[y,x] != '#':

            step.append([y,x])

            y = y + 1
        
        res_dict = {
            'y': y-1,
            'x': x,
            'len_steps': len(step),
            'nxt_dir': next_direction,
            'steps': step
        }

    if direction == 'left':

        next_direction = 'up'

        while matrix[y,x] != '#':

            step.append([y,x])

            x = x - 1
        
        res_dict = {
            'y': y,
            'x': x+1,
            'len_steps': len(step),
            'nxt_dir': next_direction,
            'steps': step
        }

    return res_dict
```

```python
steps = []

result = {
    'y': start_y,
    'x': start_x,
    'nxt_dir': 'up'
}
print(result)

i = 0

while i < 2:
    result = step_function(result['y'], result['x'], result['nxt_dir'])
    steps.extend(result['steps'])
    print(result)

    i = i + 1

print(len(steps))

#unique

unique_items = []
[unique_items.append(x) for x in steps if x not in unique_items]
print(unique_items)

len(unique_items)

```

```python
steps = []
end_points = []

result = {
    'y': start_y,
    'x': start_x,
    'nxt_dir': 'up'
}
print(result)

while [result['y'], result['x']] not in end_points:
    result = step_function(result['y'], result['x'], result['nxt_dir'])
    end_points.append([result['y'], result['x']])
    print(result)

print(end_points)

```
