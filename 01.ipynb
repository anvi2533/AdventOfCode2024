```python
reports = []

with open('Inputs\input_02.txt', 'r') as file:

    for x in file:

        string_list = x.split(' ')

        int_list = [int(y) for y in string_list]

        reports.append(int_list)

reports

```

```python
invalid_reports = []

check_list = []

for i in reports:

    report = i

    for j in list(range(len(report)-1)):

        if report[j] == report[j+1]:

            invalid_reports.append(report)

            break
    
print(f'reports: {reports}')

print(f'invalid reports: {invalid_reports}')

valid_reports = [item for item in reports if item not in invalid_reports]

print(f'valid reports: {valid_reports}')

for report in valid_reports:

    diff_list = []

    for k in list(range(len(report)-1)):

        diff = report[k+1]-report[k]

        diff_list.append(diff)

    diff_max = max(diff_list)

    diff_min = min(diff_list)

    print(diff_max, diff_min)

    if report[0] < report[1]:

        if diff_max <= 3 and diff_min > 0:

            print('safe')

            check_list.append(1)

        else:

            print('unsafe')
        
    if report[0] > report[1]:

        if diff_max < 0 and diff_min >= -3:

            print('safe')

            check_list.append(1)

        else:

            print('unsafe')

print(sum(check_list))    

```

```python
valid_reports = []

print(len(reports))

for i in reports:

    report = i

    diff_list = []

    for j in list(range(len(report)-1)):

        diff_list.append(report[j+1]-report[j])

    if diff_list[0] < 0:

        diff_list = [-x for x in diff_list]

    if 0 not in diff_list:

        max_diff = max(diff_list)

        min_diff = min(diff_list)

        if max_diff <= 3 and min_diff > 0:

            valid_reports.append(report)

print(len(valid_reports))        
```

```python
neg_list = []

pos_list = []

neu_list = []

invalid_reports = [item for item in reports if item not in valid_reports]

print(len(invalid_reports))

for i in invalid_reports:

    report = i

    diff_list = []

    for j in list(range(len(report)-1)):

        diff = report[j+1] - report[j]

        diff_list.append(diff)

    max_diff = max(diff_list)

    min_diff = min(diff_list)

    if max_diff > 3:

        pos_list.append(report)

    if min_diff < 0:

        neg_list.append(report)

    if 0 in diff_list:

        neu_list.append(report)

print(len(pos_list), len(neg_list), len(neu_list))

pot_reports = []

for report in reports:

    points = []

    if report in pos_list:

        points.append(1)

    if report in neg_list:

        points.append(1)

    if report in neu_list:

        points.append(1)

    if sum(points) == 1:

        pot_reports.append(report)

print(len(pot_reports))

for report in pot_reports:

    print(report)

    for i,item in enumerate (report):

        print(i, item)

    break






    
    
```
